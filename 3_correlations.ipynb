```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
plt.style.use('ggplot')
```

```python
data = pd.read_parquet('data/francetax_processed.parquet')

# Calculate correlation coefficients
correlation_tax_revenue = data['POPULATION'].corr(data['TAX_REVENUE'])
correlation_w_tax_revenue = data['POPULATION'].corr(data['W_TAX_REVENUE'])

print(f"Correlation between Population and Tax Revenue: {correlation_tax_revenue}")
print(f"Correlation between Population and Weighted Tax Revenue: {correlation_w_tax_revenue}")
```

```python
# Assuming 'data_complete' DataFrame contains 'POPULATION', 'TAX_REVENUE', and 'W_TAX_REVENUE' columns
correlation_matrix = data[['POPULATION', 'TAX_REVENUE', 'W_TAX_REVENUE']].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='mako', vmin=-1, vmax=1, center=0, linewidths=.5, fmt=".2f")

# Add titles and labels
plt.title('Correlation Matrix for Population, Tax Revenue, and Weighted Tax Revenue', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Display the plot
plt.show()
```

```python

# Function to format large numbers
def format_large_numbers(x, pos):
    if x >= 1e9:
        return f'{x*1e-9:.1f}B'
    elif x >= 1e6:
        return f'{x*1e-6:.1f}M'
    elif x >= 1e3:
        return f'{x*1e-3:.1f}K'
    else:
        return f'{x:.0f}'

# Create the formatter
formatter = ticker.FuncFormatter(format_large_numbers)

# Set up the scatter plot
plt.figure(figsize=(16, 6))

# Scatter plot for Population vs. Tax Revenue
plt.subplot(1, 2, 1)
plt.scatter(data['POPULATION'], data['TAX_REVENUE'], alpha=0.5)
plt.title('Population vs. Tax Revenue')
plt.xlabel('Population')
plt.ylabel('Tax Revenue')
plt.xscale('log')
plt.yscale('log')
plt.gca().xaxis.set_major_formatter(formatter)
plt.gca().yaxis.set_major_formatter(formatter)
plt.grid(True)

# Scatter plot for Population vs. Weighted Tax Revenue
plt.subplot(1, 2, 2)
plt.scatter(data['POPULATION'], data['W_TAX_REVENUE'], alpha=0.5)
plt.title('Population vs. Weighted Tax Revenue')
plt.xlabel('Population')
plt.ylabel('Weighted Tax Revenue')
plt.xscale('log')
plt.yscale('log')
plt.gca().xaxis.set_major_formatter(formatter)
plt.gca().yaxis.set_major_formatter(formatter)
plt.grid(True)

plt.tight_layout()
plt.show()
```
